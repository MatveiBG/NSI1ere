**AIDE n°1 :**

Le code ci-dessous montre comment trier un tableau à 2 dimensions selon une de ses colonnes.

```python
from operator import itemgetter #la fonction itemgetter permet de récupérer la valeur à trier pour chaque objet
liste=[[4,5,6],[1,2,3],[7,8,9]]
colonne = 1 #numéro de la colonne selon laquelle trier
liste_triee=sorted(liste, key=itemgetter(colonne)) #tri de la liste par ordre croissant selon la colonne choisie
print("Liste non triée :")
print(liste)
print("Liste triée selon la colonne ",colonne)
print(liste_triee)
```

**AIDE n°2 :**

Le code ci-dessous montre comment faire une copie d'une liste de listes.

```python
from copy import deepcopy #on a besoin de deepcopy pour faire une vraie copie de liste de listes

#la copie de liste avec la fonction list() ne fonctionne que pour des listes à une dimension
liste=[[4,5,6],[1,2,3],[7,8,9]]
liste_copie1=list(liste) #on fait une copie de liste avec la fonction list()
for i in range(3):
    liste_copie1[i].append(liste_copie1[i][1]/liste_copie1[i][2]) #on ajoute une 4ème colonne qui est le rapport de la 2ème sur la 3ème
print("Liste originale :")
print(liste) #cette liste a aussi été modifiée en même temps que la copie
print("Liste modifiée :")
print(liste_copie1) #liste_copie1 est identique à la liste originale

#la fonction deep_copie permet de faire une vraie copie de liste de listes
liste=[[4,5,6],[1,2,3],[7,8,9]]
liste_copie2=deepcopy(liste) #on crée une vraie copie de liste
for i in range(3):
    liste_copie2[i].append(liste_copie2[i][1]/liste_copie2[i][2]) #on ajoute une 4ème colonne qui est le rapport de la 2ème sur la 3ème
print("Liste originale :")
print(liste) #cette fois, la liste originale n'est pas modifiée en même temps que la copie
print("Liste modifiée :")
print(liste_copie2) #liste_copie2 est une vraie copie distincte de la liste originale
```
