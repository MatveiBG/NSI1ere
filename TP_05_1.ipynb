# <span style="color:red;"><center>TP - Créer une page web sans CSS </center></span>

## 1. Objectif du TP

Créer une page web simple, sans CSS, comportant des liens, des images, ... et une structure de base *header*, *section*, *footer*.

Evidemment, sans CSS, ce ne sera pas très joli ...

## 2. Consignes

Ce TP est très directement inspiré du travail de Eric Hartung disponible ici :

[TP : Ma première page web](http://mathartung.xyz/isn/TP/TPhtml/tp1.html)

Le but est de reconstruire une page similaire à celle-ci :


Pour cela, tu trouveras les fichiers nécessaires (images et texte) dans le répertoire *TPhtml* du dépot Github ou sur le lien [suivant](https://drive.google.com/file/d/0ByE6TdP1IvIWc1BudDVTUC1nX1U/view?usp=sharing).


### Marche à suivre :

1. Dans ton répertoire personnel, créer un fichier *index.html* dans un répertoire de son choix, nommé par exemple *TP5_1*.
2. Créer un sous répertoire *sources* dans lequel on téléchargera les 4 images disponibles ci-dessus.
3. Ouvrir le fichier *index.html* avec [Notepad++](https://notepad-plus-plus.org/downloads/v7.9.5/) ou autre éditeur de texte de son choix.
4. Copier dans ce fichier le code HTML minimal suivant :

```
<!DOCTYPE html>
<html lang="fr">
<head>
	<meta charset="UTF-8"/>
	<title>titre de la page</title>
</head>
	<body>
	

	</body>
</html>
```
5. Enregistrer son fichier HTML et l'ouvrir avec son navigateur préféré qui devrait afficher une page blanche nommée *titre de la page*.
6. Compléter le fichier HTML en le sauvegardant régulièrement et en rafraichissant l'affichage du navigateur pour constater les évolutions.
    - Changer le titre de la page en "Premier site web au monde".
    - Insérer le texte de base entre les balises *body*.
    - Ajouter les balises structurantes pour le *header*, la *section* principale et le *footer*.
    - Compléter le *header* en y insérant l'image *CERNLogo.gif* qui pointe vers le site *http://info.cern.ch*.
    - Définir les phrases "Bienvenue sur ...", "Le site Web du tout ..." et "1990, une année ..." comme des titres de niveaux 2, 3 et 4.
    - Définir les 3 phrases commencant par "En février", "En avril" et "En octobre" comme une liste non numérotée.
    - Scinder le reste du textes en paragraphes, comme sur l'image.
    - Utiliser la balise ``<abbr>`` pour définir sous forme d'info-bulles au survol de la souris les acronymes CERN(Laboratoire européen pour la recherche nucléaire) et SLAC (Centre de l'accélérateur linéaire de Stanford). Il faut pour cela utiliser l'attribut *title*.
    - Insérer les 3 images restantes en les intégrant dans des balises ``<figure>`` et en leur ajoutant une légende grâce à la balise ``<figcaption>``. Le texte des légendes est à la fin du fichier *texte de base.txt*.
    - Appliquer la balise ``<code>`` à l'URL *http://info.cern.ch/hypertext/WWW/TheProject.html*.
    - Ajouter les liens suivants :
    |Texte du lien|URL du lien|
    |:-:|:-|
    |proposition|http://info.cern.ch/Proposal-fr.html|
    |prototype|http://info.cern.ch/NextBrowser-fr.html|
    |cette page|http://www.w3.org/History/19921103-hypertext/hypertext/WWW/TheProject.html|
    |navigateur|http://info.cern.ch/LMBrowser-fr.html|
    - Ajouter un *lien email* dans le *footer* grâce à l'URL *mailto:web.communications@cern.ch*.
